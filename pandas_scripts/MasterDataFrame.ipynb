## Read Mulitple Data Frames

```python
df1 = pd.read_csv('/Users/newuser/Dropbox/Datamining/Crypto-Pirates/Datasets/bitcoin_price.csv')
df2 = pd.read_csv('/Users/newuser/Dropbox/Datamining/Crypto-Pirates/Datasets/monero_price.csv')
df3 = pd.read_csv('/Users/newuser/Dropbox/Datamining/Crypto-Pirates/Datasets/litecoin_price.csv')
```

# Concatenating Multiple Data Frames

Before I concatenate multiple data frames, I first added a new column to each data frame which is going to be used for map reducing later.

```python
df1['CC'] = 'Bitcoin'
df2['CC'] = 'Monero'
df3['CC'] = 'Litecoin'
master_df = pd.concat([df1,df2,df3])
master_df
```

# Map Reduce

```python
master_df.groupby('CC').mean()
```

```python
master_df.groupby('CC').min()
```

```python
master_df.groupby('CC').max()
```
