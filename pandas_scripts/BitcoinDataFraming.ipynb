# Finding Simple Moving Average

We want to plot a simple moving average in order to discover crossovers. Crossovers are good indicators for buying and selling.

A simple moving average (SMA) is an arithmetic moving average calculated by adding the closing price of the security for a number of time periods and then dividing this total by the number of time periods.

Read more: Simple Moving Average (SMA) https://www.investopedia.com/terms/s/sma.asp#ixzz50bHYsA5m 
Follow us: Investopedia on Facebook

First, find the overall SMA of our dataframe

```python
# This is a constant horizontal value
close = btc_df['Close']
day1 = btc_df['Date'].min()
lastDay = btc_df['Date'].max()
sma = close.mean()
sma
```

```python
day1
```

```python
lastDay
```

```python
# This is a moving average curve
btc_df['MA'] = btc_df['Close'].rolling(5).mean()
btc_df
```

# Visual of MA and Close

```python
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
trace1 = go.Scatter(
    x = btc_df['Date'],
    y = btc_df['Close'],
    mode = 'Lines',
    name = 'Bitcoin Close Price'
)
trace2 = go.Scatter(
    x = btc_df['Date'],
    y = btc_df['MA'],
    mode = 'Lines',
    name = 'Bitcoin Moving Average'
)

data = [trace1,trace2]

py.iplot(data, filename='basic-scatter')
```
