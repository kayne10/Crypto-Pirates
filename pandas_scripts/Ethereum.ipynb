```python
import numpy as np
import pandas as pd
```

```python
eth_df = pd.read_csv('/Users/troy/Github/courses/crypto-pirates/cryptocurrencypricehistory/ethereum_price.csv')
new_index = eth_df['Date']
eth_df.head()
```

### Adding Columns

```python
eth_df['Daily Range'] = eth_df['High'] - eth_df['Low']
eth_df.head()
```

```python
eth_df['MA'] = eth_df['Close'].rolling(5).mean()
eth_df
```

### Adding the mean

```python
eth_df['Daily Range'].mean()
```

```python
'''
1. Add new column. 2. Define Conditions. 3. Change necessary value to the column that meets the condition.
'''

eth_df['Significant'] = ''

cond1 = eth_df['Daily Range'] > eth_df['Daily Range'].mean()
cond2 = eth_df['Daily Range'] < eth_df['Daily Range'].mean()

yes = eth_df[cond1]
no = eth_df[cond2]

column = 'Significant'
eth_df.loc[cond1, column] = 'Yes'
eth_df.loc[cond2, column] = 'No'
eth_df.set_index('Significant', inplace=True)
eth_df
```

# Visual of MA and CLose

```python
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
trace1 = go.Scatter(
    x = eth_df['Date'],
    y = eth_df['Close'],
    mode = 'Lines',
    name = 'Ethereum Close Price'
)
trace2 = go.Scatter(
    x = eth_df['Date'],
    y = eth_df['MA'],
    mode = 'Lines',
    name = 'Ethereum Moving Average'
)

data = [trace1,trace2]

py.iplot(data, filename='basic-scatter')
```
